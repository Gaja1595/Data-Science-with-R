```r
library("caTools")
df=read.csv("USA_Housing.csv")
df1=subset(df,select=-c(Address))
df1
```

```r
s=sample.split(df1,SplitRatio=0.7)
s
```

```r
train=subset(df1,s==TRUE)
test=subset(df1,s==FALSE)
test
```

```r
mod=lm(Price~.,data=train)
```

```r
ref=predict(mod,test)
```

```r
ref
```

```r
df2=data.frame(test$Price,ref)
df2
```

```r
r2=summary(mod)
r2
```
