```r
df=data.frame(dose=c("D0.5","D1","D2"),
              len=c(4.2,10,29.5))
head(df)
```

```r
install.packages("tidyverse",type="binary")
```

```r
library(ggplot2)
             #basic barplot
p=ggplot(data=df,aes(x=dose,y=len))+
geom_bar(stat="identity",color="blue",fill="red",width=0.3)
p



p+coord_flip()  #horizontal bar plot#basic barplot
     
```

```r
p=ggplot(data=df,aes(x=dose,y=len))+
geom_bar(stat="identity",width=0.3)+
theme_grey()
p
```

```r
p=ggplot(data=df,aes(x=dose,y=len))+
geom_bar(stat="identity",width=0.3)+
theme_dark()
p
```

```r
p=ggplot(data=df,aes(x=dose,y=len))+
geom_bar(stat="identity",width=0.3)+
theme_test()
p
```
